## Marker overlap
Next, we are analyzing the overlap of marker genes between the cell types of the two different studies per tissue.
For this, we are using `sc.tl.rank_genes_groups`.

```python
def markerIntersect(adata):
    ret = {}
    for n in adata.obs.tissue.unique():
        print(n)
        tmp = adata[adata.obs.tissue==n]
        split = scIB.utils.splitBatches(tmp, 'study')
        markers = []
        for i in split:
            print(i.obs.study.unique())
            sc.pp.filter_genes(i, min_counts=.5)
            sc.tl.rank_genes_groups(i, groupby='cell_type', key_added='rank_genes')
            markers.append(pd.DataFrame.from_records(i.uns['rank_genes']['names']))
        if len(split)!= 3:
            markers.append(markers[1])
        out1 = pd.DataFrame(columns=markers[0].columns)
        out2 = pd.DataFrame(columns=markers[0].columns)
        for i in markers[0]:
            a = set(markers[0][i])
            for j in markers[1]:
                b = set(markers[1][j])
                out1.loc[j,i]=len(a.intersection(b))
            for j in markers[2]:
                b = set(markers[2][j])
                out2.loc[j,i]=len(a.intersection(b))
        ret[n]=out1.add(out2).fillna(out1*2).fillna(out2*2).astype(int)
    return ret
```

```python
sc.pp.filter_genes(adata, min_counts=10)
```

```python
intersect_markers = markerIntersect(adata)
```
